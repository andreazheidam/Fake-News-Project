```python
import pandas as pd

df = pd.read_csv("995K_subset.csv", dtype={0: str, 1: str})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

missing_values = df.isnull()

plt.figure(figsize=(10, 6))
sns.heatmap(missing_values, cmap=['hotpink', 'slateblue'], cbar=False)
plt.title('Presence of Missing Values')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.show()
```
